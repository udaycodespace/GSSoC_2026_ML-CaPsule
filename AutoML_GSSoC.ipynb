Regression with H2O AutoML

Dataset - Boston_housing

```python
import os
import pandas as pd
import h2o
from h2o.automl import H2OAutoML
import requests
import matplotlib.pyplot as plt
import seaborn as sns

# Download the dataset
url = "http://lib.stat.cmu.edu/datasets/boston"
response = requests.get(url)
data = response.text.split("\n")

# Process the dataset
header = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
raw_data = data[22:]  # The actual data starts at line 23

# Parsing the data into a format readable by pandas
parsed_data = []
for i in range(0, len(raw_data) - 1, 2):
    row1 = list(map(float, raw_data[i].strip().split()))
    row2 = list(map(float, raw_data[i+1].strip().split()))
    parsed_data.append(row1 + row2)

# Create a DataFrame
df = pd.DataFrame(parsed_data, columns=header)

# Initialize H2O (Using minimal resource allocation threads to prevent CPU clashing)
h2o.init(nthreads=2, max_mem_size="2G")

# Convert to H2OFrame
hf = h2o.H2OFrame(df)

# Define target and features
y = 'MEDV'
x = hf.columns
x.remove(y)

# Split the data into train and test sets
train, test = hf.split_frame(ratios=[.8], seed=1234)

max_seconds = 15          # Absolute strict time limit
max_models = 2            # Train exactly two models and stop
exclude_algos = ["DeepLearning", "StackedEnsemble"] # Skip time-consuming algorithms

# Run optimized H2O AutoML
aml = H2OAutoML(
    max_runtime_secs=max_seconds,
    max_models=max_models,
    exclude_algos=exclude_algos,
    seed=1234
)
aml.train(x=x, y=y, training_frame=train)

# View the AutoML Leaderboard
lb = aml.leaderboard
print(lb)

# Predict on the test set
preds = aml.leader.predict(test)
print(preds.head())
```

```python
# Convert predictions to a pandas DataFrame
preds_df = preds.as_data_frame()
test_df = test.as_data_frame()

# Scatter plot of actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(test_df['MEDV'], preds_df['predict'], alpha=0.7)
plt.xlabel('Actual MEDV')
plt.ylabel('Predicted MEDV')
plt.title('Actual vs Predicted MEDV')
plt.plot([test_df['MEDV'].min(), test_df['MEDV'].max()], [test_df['MEDV'].min(), test_df['MEDV'].max()], 'k--', lw=3)
plt.show()

# Residuals plot
residuals = test_df['MEDV'] - preds_df['predict']
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True)
plt.xlabel('Residuals')
plt.title('Distribution of Residuals')
plt.show()
```

Classification with H2O AutoML

Dataset - Iris

```python
import os
import h2o
from h2o.automl import H2OAutoML
from sklearn.datasets import load_iris
import pandas as pd

# 1. Initialize H2O (Capped resources to prevent GitHub worker memory locks)
h2o.init(nthreads=1, max_mem_size="2G")

# 2. Load dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# 3. Convert to H2OFrame
hf = h2o.H2OFrame(df)

# 4. Define target and features
y = 'species'
x = hf.columns
x.remove(y)

# 5. Split the data into train and test sets
train, test = hf.split_frame(ratios=[.8], seed=1234)
max_seconds = 10           # Drops time window from 300s to 10s
max_models = 1             # Stops execution immediately after the first model finishes
exclude_algos = ["DeepLearning", "StackedEnsemble", "XGBoost", "GBM"]  # Only allows fast GLM/DRF

# 7. Run optimized H2O AutoML
aml = H2OAutoML(
    max_runtime_secs=max_seconds,
    max_models=max_models,
    exclude_algos=exclude_algos,
    seed=1234
)
aml.train(x=x, y=y, training_frame=train)

# 8. View the AutoML Leaderboard
lb = aml.leaderboard
print(lb)

# 9. Predict on the test set
preds = aml.leader.predict(test)
print(preds.head())
```

```python
# Convert predictions to pandas DataFrame
preds_df = preds.as_data_frame()
test_df = test.as_data_frame()

# Confusion matrix
confusion_matrix = pd.crosstab(test_df['species'], preds_df['predict'], rownames=['Actual'], colnames=['Predicted'])
print(confusion_matrix)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ROC curve (for binary classification tasks)
if len(iris.target_names) == 2:
    perf = aml.leader.model_performance(test_data=test)
    perf.plot()

# Shutdown H2O
h2o.shutdown()
```

Regression with TPOT

Dataset - California_housing

```python
from tpot import TPOTRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score # Safe metric replacement
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load data
california = fetch_california_housing()
X, y = california.data, california.target

# 2. Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# 3. CRITICAL FOR SPEED: Subsample data to a quick 10% slice
X_train_fast, _, y_train_fast, _ = train_test_split(
    X_train, y_train, train_size=0.1, random_state=123
)

# 4. Pure Clean Initialization (Strictly single-threaded for Colab RAM safety)
tpot = TPOTRegressor(
    generations=0,
    population_size=2,
    random_state=123,
    n_jobs=1,                                 # Threading conflict is resolved by keeping this 1
    cv=2,
    max_time_mins=1.5
)

# 5. Execute directly (Removed the broken parallel_backend context wrapper)
print("Beginning pipeline search...")
tpot.fit(X_train_fast, y_train_fast)

# 6. FIX: Extract the model with an explicit safety check to prevent AttributeErrors
if hasattr(tpot, 'fitted_pipeline_') and tpot.fitted_pipeline_ is not None:
    best_model = tpot.fitted_pipeline_
    preds = best_model.predict(X_test)

    # Calculate and print the final R2 score
    test_r2 = r2_score(y_test, preds)
    print("\nR2 score on test set:", test_r2)

    # 7. Safe Export Strategy
    try:
        tpot.export('tpot_california_pipeline.py')
        print("Pipeline script saved successfully via native exporter!")
    except AttributeError:
        # Fallback file writer if the wrapper method is missing
        with open('tpot_california_pipeline.py', 'w') as f:
            f.write(f"from sklearn.pipeline import Pipeline\nexported_pipeline = {repr(best_model)}")
        print("Pipeline script saved successfully via backup exporter!")

    # 8. Plot 1: Actual vs Predicted values
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(y_test, preds, alpha=0.6, color='teal')
    plt.xlabel('Actual House Price')
    plt.ylabel('Predicted House Price')
    plt.title('Actual vs Predicted House Price (TPOT)')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)

    # 9. Plot 2: Residuals distribution plot
    plt.subplot(1, 2, 2)
    residuals = y_test - preds
    sns.histplot(residuals, kde=True, color='crimson')
    plt.xlabel('Residuals')
    plt.title('Distribution of Residuals (TPOT)')

    plt.tight_layout()
    plt.show()

else:
    print("\n[CRITICAL ERROR]: TPOT was terminated by Colab before completing a single generation.")
    print("Please go to 'Runtime' -> 'Restart session' in the top menu to clear stuck cache, then run this cell again.")
```

Classification using TPOT

Dataset - Iris

```python
from tpot import TPOTClassifier
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report # Classification metrics
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load data
california = fetch_california_housing()
X, y_continuous = california.data, california.target

# Convert continuous prices into binary classification labels (1: Above Median, 0: Below Median)
median_price = np.median(y_continuous)
y = (y_continuous > median_price).astype(int)

# 2. Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# 3. CRITICAL FOR SPEED: Subsample data to a quick 10% slice
X_train_fast, _, y_train_fast, _ = train_test_split(
    X_train, y_train, train_size=0.1, random_state=123
)

# 4. Pure Clean Initialization (Strictly single-threaded for Colab RAM safety)
tpot = TPOTClassifier(
    generations=0,
    population_size=2,
    random_state=123,
    n_jobs=1,                                 # Threading conflict is resolved by keeping this 1
    cv=2,
    max_time_mins=1.5
)

# 5. Execute directly
print("Beginning classifier pipeline search...")
tpot.fit(X_train_fast, y_train_fast)

# 6. FIX: Extract the model with an explicit safety check to prevent AttributeErrors
if hasattr(tpot, 'fitted_pipeline_') and tpot.fitted_pipeline_ is not None:
    best_model = tpot.fitted_pipeline_
    preds = best_model.predict(X_test)

    # Calculate and print the final Accuracy score and report
    test_accuracy = accuracy_score(y_test, preds)
    print("\nAccuracy score on test set:", test_accuracy)
    print("\nClassification Report:\n", classification_report(y_test, preds))

    # 7. Safe Export Strategy
    try:
        tpot.export('tpot_california_classifier_pipeline.py')
        print("Pipeline script saved successfully via native exporter!")
    except AttributeError:
        # Fallback file writer if the wrapper method is missing
        with open('tpot_california_classifier_pipeline.py', 'w') as f:
            f.write(f"from sklearn.pipeline import Pipeline\nexported_pipeline = {repr(best_model)}")
        print("Pipeline script saved successfully via backup exporter!")

    # 8. Plot 1: Confusion Matrix
    from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    cm = confusion_matrix(y_test, preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Below Median', 'Above Median'])
    disp.plot(ax=plt.gca(), cmap='Blues', values_format='d')
    plt.title('Confusion Matrix (TPOT)')

    # 9. Plot 2: Predicted Class Probabilities Distribution
    plt.subplot(1, 2, 2)
    if hasattr(best_model, "predict_proba"):
        probs = best_model.predict_proba(X_test)[:, 1]
        sns.histplot(probs, kde=True, color='purple', bins=20)
        plt.xlabel('Predicted Probability of Being Above Median')
        plt.title('Distribution of Predicted Probabilities')
    else:
        plt.text(0.5, 0.5, 'Selected model does not support\nprobability estimations',
                 ha='center', va='center', fontsize=12, color='red')
        plt.title('Probability Distribution (Not Available)')

    plt.tight_layout()
    plt.show()

else:
    print("\n[CRITICAL ERROR]: TPOT was terminated by Colab before completing a single generation.")
    print("Please go to 'Runtime' -> 'Restart session' in the top menu to clear stuck cache, then run this cell again.")
```
