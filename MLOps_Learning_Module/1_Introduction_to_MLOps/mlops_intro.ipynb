# Introduction to MLOps - Practical Examples
## Understanding MLOps with Real-World Scenarios

## Example 1: Without MLOps (Manual Workflow)

```python
# Scenario: Spam Detector WITHOUT MLOps
# Problem: You trained a model, now what?

import pickle
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Simulate training a model
print("=== WITHOUT MLOPS WORKFLOW ===")
print()

# Step 1: Train model (manually)
X, y = make_classification(n_samples=1000, n_features=20, n_informative=15)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print(f"Step 1: Model trained. Accuracy: {accuracy:.2%}")
print()

# Step 2: Save model to file (manually named)
with open('model_final.pkl', 'wb') as f:
    pickle.dump(model, f)
print("Step 2: Model saved to 'model_final.pkl'")
print()

# Problem: Version control nightmare
print("Problems WITHOUT MLOps:")
print("❌ Which model version is best?")
print("❌ When was it trained?")
print("❌ What parameters were used?")
print("❌ How does it compare to other models?")
print("❌ No one knows which version is in production!")
print("❌ No monitoring - model performance degrades silently")
print("❌ Manual retraining process - error prone")
print()
print("Result: Model sits in production, performance drops over time, no one notices!")
```

## Example 2: With MLOps (Organized Workflow)

```python
# Scenario: Same model BUT with MLOps practices

from datetime import datetime
import json

print("\n=== WITH MLOPS WORKFLOW ===")
print()

# Simulate MLOps tracking
mlops_metadata = {
    'model_name': 'spam_detector',
    'version': 'v1.0',
    'timestamp': datetime.now().isoformat(),
    'algorithm': 'Random Forest',
    'parameters': {
        'n_estimators': 100,
        'max_depth': 15,
        'random_state': 42
    },
    'metrics': {
        'accuracy': accuracy,
        'precision': 0.95,
        'recall': 0.93,
        'f1_score': 0.94
    },
    'data_info': {
        'training_samples': len(X_train),
        'test_samples': len(X_test),
        'features': X.shape[1]
    },
    'status': 'ready_for_deployment'
}

# Save metadata
with open('spam_detector_v1.0_metadata.json', 'w') as f:
    json.dump(mlops_metadata, f, indent=2)

print("MLOps Workflow:")
print(f"✓ Model: {mlops_metadata['model_name']}")
print(f"✓ Version: {mlops_metadata['version']}")
print(f"✓ Trained: {mlops_metadata['timestamp']}")
print(f"✓ Algorithm: {mlops_metadata['algorithm']}")
print(f"✓ Accuracy: {mlops_metadata['metrics']['accuracy']:.2%}")
print(f"✓ Status: {mlops_metadata['status']}")
print()

print("Benefits WITH MLOps:")
print("✓ Clear version tracking")
print("✓ All metadata recorded")
print("✓ Easy comparison between models")
print("✓ Reproducible results")
print("✓ Automatic monitoring")
print("✓ Automated retraining")
print("✓ Team collaboration")
print()
print("Result: Clear visibility, automated processes, production-ready!")
```

## Example 3: MLOps Key Concepts

```python
# Concept 1: Deployment
print("\n=== CONCEPT 1: DEPLOYMENT ===")
print()
print("Before deployment:")
print("  Model → Your Computer → Only you can use it")
print()
print("After deployment:")
print("  Model → Server → API → Users can access")
print()
print("Example: Spam Detector API")
print("-" * 50)
print("""
POST /predict
Input: {"email_text": "Free money! Click here!"}
Output: {"prediction": 1, "probability": 0.98}  # 1 = Spam
""")
```

```python
# Concept 2: Experiment Tracking
print("\n=== CONCEPT 2: EXPERIMENT TRACKING ===")
print()

experiments = [
    {'name': 'exp_1', 'algorithm': 'Logistic Regression', 'accuracy': 0.92},
    {'name': 'exp_2', 'algorithm': 'SVM', 'accuracy': 0.94},
    {'name': 'exp_3', 'algorithm': 'Random Forest', 'accuracy': 0.96},
    {'name': 'exp_4', 'algorithm': 'Gradient Boosting', 'accuracy': 0.97},
]

print("Experiment Comparison:")
print("-" * 50)
for exp in experiments:
    print(f"{exp['name']:10} | {exp['algorithm']:18} | Accuracy: {exp['accuracy']:.2%}")

best_exp = max(experiments, key=lambda x: x['accuracy'])
print("-" * 50)
print(f"BEST: {best_exp['name']} - {best_exp['algorithm']} ({best_exp['accuracy']:.2%})")
print()
print("Without tracking: ❌ We'd forget which experiment was best")
print("With tracking: ✓ Clear winner - Gradient Boosting (97%)")
```

```python
# Concept 3: Monitoring
print("\n=== CONCEPT 3: MONITORING ===")
print()

# Simulated monitoring over 90 days
import matplotlib.pyplot as plt

days = list(range(1, 91, 10))
accuracy_over_time = [0.97, 0.96, 0.95, 0.93, 0.91, 0.88, 0.85, 0.82, 0.80]

print("Model Performance Over Time:")
print("-" * 50)
for day, acc in zip(days, accuracy_over_time):
    status = "✓" if acc >= 0.90 else "⚠" if acc >= 0.85 else "❌"
    print(f"Day {day:2d}: Accuracy {acc:.2%} {status}")

print("-" * 50)
print("\nObservation: Accuracy drops from 97% to 80% in 90 days!")
print("Action: Retrain model with new data")
print()

# Visualization
plt.figure(figsize=(10, 5))
plt.plot(days, accuracy_over_time, marker='o', linewidth=2, markersize=8)
plt.axhline(y=0.90, color='r', linestyle='--', label='Alert threshold (90%)')
plt.xlabel('Days')
plt.ylabel('Model Accuracy')
plt.title('Model Performance Degradation Over Time')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print("Without monitoring: ❌ Accuracy drops silently, users frustrated")
print("With monitoring: ✓ Alert triggered, model retrained automatically")
```

```python
# Concept 4: Versioning
print("\n=== CONCEPT 4: MODEL VERSIONING ===")
print()

versions = [
    {'version': 'v1.0', 'accuracy': 0.94, 'status': 'retired'},
    {'version': 'v1.1', 'accuracy': 0.95, 'status': 'retired'},
    {'version': 'v2.0', 'accuracy': 0.96, 'status': 'production'},
    {'version': 'v2.1', 'accuracy': 0.97, 'status': 'staging'},
]

print("Model Versions:")
print("-" * 60)
for v in versions:
    print(f"{v['version']} | Accuracy: {v['accuracy']:.2%} | Status: {v['status']:11}")
print("-" * 60)

print()
print("Benefits of versioning:")
print("✓ Easy rollback if new version has issues")
print("✓ A/B testing between versions")
print("✓ Track improvement over time")
print("✓ Know which version is in production")
```

```python
# Concept 5: Automation
print("\n=== CONCEPT 5: AUTOMATION ===")
print()

print("Manual Process (❌ Error-prone, time-consuming):")
print("  1. Developer writes code")
print("  2. Manually test locally")
print("  3. Manually merge code")
print("  4. Manual deployment to production")
print("  5. Manual monitoring")
print("  6. Manual retraining when needed")
print()

print("Automated Process (✓ Efficient, consistent):")
print("  1. Developer writes code")
print("  2. Automated tests run")
print("  3. Automated code merge")
print("  4. Automated deployment")
print("  5. Automated monitoring")
print("  6. Automated retraining")
print()

print("Time saved with automation:")
print(f"  Manual: 40 hours/week")
print(f"  Automated: 5 hours/week")
print(f"  Savings: 35 hours/week! 🚀")
```
