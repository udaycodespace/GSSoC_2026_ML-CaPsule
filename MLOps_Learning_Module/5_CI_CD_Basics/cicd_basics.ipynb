## Part 1: Understanding CI/CD

```python
print("=== UNDERSTANDING CI/CD ===")
print()

print("Without CI/CD (Manual Process):")
print("-" * 50)
manual = [
    "1. Developer writes code",
    "2. Manually test locally",
    "3. Email code to team",
    "4. Team manually tests",
    "5. Manual code review",
    "6. Manual merge to main",
    "7. Manual deployment",
    "8. Issues found in production",
    "9. Manual debugging",
    "10. Manual rollback"
]

for step in manual:
    print(step)

print()
print("Problems:")
print("❌ Slow process (days to deploy)")
print("❌ Error-prone (manual steps)")
print("❌ Issues in production")
print("❌ Difficult to rollback")
print("❌ Team bottleneck")
print()
print()

print("With CI/CD (Automated Process):")
print("-" * 50)
automated = [
    "1. Developer pushes code to GitHub",
    "2. Automated tests run",
    "3. Code quality checks",
    "4. Model accuracy tests",
    "5. If all pass → Auto deploy",
    "6. If fail → Notify developer",
    "7. Rollback automated if needed"
]

for step in automated:
    print(step)

print()
print("Benefits:")
print("✓ Fast deployment (minutes)")
print("✓ Consistent process")
print("✓ Early issue detection")
print("✓ Easy rollback")
print("✓ Team efficiency")
```

## Part 2: GitHub Actions Workflow Example

```python
print("\n=== GITHUB ACTIONS WORKFLOW ===")
print()

workflow_yaml = '''
name: ML Pipeline

on:
  push:
    branches: [ main, develop ]
  pull_request:
    branches: [ main ]

jobs:
  test:
    runs-on: ubuntu-latest
    
    steps:
      # Step 1: Get code from GitHub
      - uses: actions/checkout@v2
      
      # Step 2: Set up Python
      - uses: actions/setup-python@v2
        with:
          python-version: '3.9'
      
      # Step 3: Install dependencies
      - name: Install dependencies
        run: |
          pip install -r requirements.txt
      
      # Step 4: Run lint check
      - name: Lint code
        run: |
          pip install pylint
          pylint src/*.py
      
      # Step 5: Run unit tests
      - name: Run unit tests
        run: |
          pip install pytest
          pytest tests/test_preprocessing.py
      
      # Step 6: Test model predictions
      - name: Test model
        run: |
          pytest tests/test_model.py
      
      # Step 7: Check accuracy threshold
      - name: Verify model accuracy
        run: |
          pytest tests/test_accuracy.py
      
      # Step 8: Deploy if all pass
      - name: Deploy
        if: success()
        run: |
          python deploy.py
'''

print("File: .github/workflows/ml_pipeline.yml")
print()
print(workflow_yaml)

print("\nWhat this workflow does:")
print("1. Runs on every push to main or develop")
print("2. Runs on every pull request")
print("3. Automatically tests code")
print("4. Automatically checks code quality")
print("5. Automatically tests model")
print("6. Automatically deploys if all pass")
print("7. Notifies if any step fails")
```

## Part 4: Running Tests Locally

```python
print("\n=== RUNNING TESTS LOCALLY ===")
print()

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

print("Setting up test data...")
X, y = make_classification(n_samples=1000, n_features=20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

print("Running tests:")
print("-" * 50)

# Test 1: Prediction shape
print("\nTest 1: Prediction shape")
predictions = model.predict(X_test)
assert predictions.shape == (len(X_test),)
print("✓ PASSED")

# Test 2: Prediction range
print("\nTest 2: Prediction range (0-1)")
assert all(pred in [0, 1] for pred in predictions)
print("✓ PASSED")

# Test 3: Accuracy threshold
print("\nTest 3: Model accuracy > 80%")
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy:.2%}")
assert accuracy > 0.80
print("✓ PASSED")

# Test 4: Model reproducibility
print("\nTest 4: Model reproducibility")
model2 = RandomForestClassifier(n_estimators=100, random_state=42)
model2.fit(X_train, y_train)
acc1 = model.score(X_test, y_test)
acc2 = model2.score(X_test, y_test)
assert acc1 == acc2
print("✓ PASSED")

print()
print("-" * 50)
print("All tests passed! ✓")
print()
print("To run actual pytest tests:")
print("  pytest tests/")
print("  pytest tests/test_model.py -v")
print("  pytest tests/ -k accuracy")
```
