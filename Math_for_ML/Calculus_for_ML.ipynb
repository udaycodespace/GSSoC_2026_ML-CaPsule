# Calculus for Machine Learning

Calculus, specifically differential calculus, is how ML models learn. By understanding how a function changes (derivatives), models can iteratively adjust their parameters to minimize error.

## 1. Derivatives and Gradients

- **Derivative**: The rate of change of a function. Tells us the slope at a specific point.
- **Partial Derivative**: How a multi-variable function changes if we tweak just one variable, keeping others constant.
- **Gradient**: A vector containing all the partial derivatives of a function. It points in the direction of the steepest ascent.

```python
import numpy as np
import matplotlib.pyplot as plt

# A simple loss function: L(w) = w^2
w = np.linspace(-10, 10, 100)
L = w**2

plt.plot(w, L)
plt.title('Loss Function: $L(w) = w^2$')
plt.xlabel('Weight (w)')
plt.ylabel('Loss (L)')
plt.grid()
# Show slope at w=5
plt.plot(5, 25, 'ro')
plt.annotate('Slope = 10', xy=(5, 25), xytext=(2, 50), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

## 2. The Chain Rule

When functions are nested (like a neural network with multiple layers), we use the chain rule to calculate the derivative of the output with respect to the input.

$\frac{dz}{dx} = \frac{dz}{dy} \cdot \frac{dy}{dx}$

## 🧠 Connection to Machine Learning

### Gradient Descent
Gradient Descent is the core optimization algorithm in ML. It uses the gradient to update weights to minimize the loss function $L$.

The update rule relies heavily on the partial derivative of the loss with respect to a weight $w$:

$$ w_{new} = w_{old} - \alpha \frac{\partial L}{\partial w} $$

Where $\alpha$ is the learning rate.

### Backpropagation
In neural networks, the loss depends on the output of the final layer, which depends on the previous layer, and so on. The **Chain Rule** is the mathematical engine of backpropagation, allowing us to calculate $\frac{\partial L}{\partial w}$ for weights deep within the network.
