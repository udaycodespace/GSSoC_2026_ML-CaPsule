# Linear Algebra for Machine Learning

Linear algebra is the foundational math that makes Machine Learning (especially Deep Learning) possible. Data is represented as vectors and matrices, and algorithms use matrix operations to process this data quickly.

## 1. Vectors and Matrices

- **Vector**: A 1D array of numbers. Represents features of a single data point.
- **Matrix**: A 2D array of numbers. Represents a dataset (rows = samples, cols = features) or network weights.

```python
import numpy as np
import matplotlib.pyplot as plt

# Vector
v = np.array([1, 2, 3])
print('Vector v:', v)

# Matrix
M = np.array([[1, 2], [3, 4], [5, 6]])
print('\nMatrix M:\n', M)
```

### Visualizing Vectors
Let's visualize a 2D vector.

```python
plt.figure(figsize=(4,4))
plt.quiver(0, 0, 2, 3, angles='xy', scale_units='xy', scale=1, color='b')
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid()
plt.title('Vector [2, 3]')
plt.show()
```

## 2. Dot Product

The dot product of two vectors tells us how much they point in the same direction. It's fundamental to measuring similarity and calculating weighted sums in neural networks.

```python
a = np.array([2, 3])
b = np.array([1, 4])
dot_product = np.dot(a, b)
print('Dot product of a and b:', dot_product)
```

## 3. Matrix Multiplication and Transpose

- **Matrix Multiplication**: Used to pass data through network layers.
- **Transpose**: Flipping a matrix over its diagonal, turning rows into columns. Often needed to align dimensions for multiplication.

```python
W = np.array([[0.1, 0.2], [0.3, 0.4]]) # Weights
x = np.array([1.0, 2.0]) # Input

output = np.dot(W, x)
print('Output (W * x):', output)

print('\nTranspose of W:\n', W.T)
```

## 🧠 Connection to Machine Learning

### Linear Regression
In scikit-learn's `LinearRegression`, the model learns a vector of weights $\vec{w}$. To make a prediction for a new input vector $\vec{x}$, it simply computes the dot product: $\hat{y} = \vec{w} \cdot \vec{x} + b$.

### Neural Networks
A layer in a neural network performs a matrix multiplication between the weight matrix $W$ and the input vector $x$, followed by an activation function: $y = \sigma(W x + b)$. The transpose operation is heavily used during backpropagation to distribute gradients back through the network.
