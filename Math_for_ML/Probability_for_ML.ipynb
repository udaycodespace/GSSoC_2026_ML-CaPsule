# Probability & Distributions for ML

Probability theory provides a framework for modeling uncertainty. In ML, data is often noisy, and models make probabilistic predictions rather than absolute certainties.

## 1. Probability Basics & Bayes Theorem

- **Conditional Probability**: The probability of event A given that event B has occurred, denoted as $P(A|B)$.
- **Bayes' Theorem**: A fundamental theorem for updating probabilities based on new evidence.

$$ P(A\mid B)=\frac{P(B\mid A)P(A)}{P(B)} $$

## 2. Normal Distribution (Gaussian)

The Normal Distribution is the most important probability distribution in statistics, forming the famous 'bell curve'. Many ML algorithms assume data follows this distribution.

- **PDF (Probability Density Function)**: Shows the relative likelihood of a continuous variable taking a specific value.
- **CDF (Cumulative Distribution Function)**: Shows the probability that a variable will be less than or equal to a specific value.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-5, 5, 100)
# PDF (mean=0, std=1)
pdf = norm.pdf(x, 0, 1)
# CDF
cdf = norm.cdf(x, 0, 1)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].plot(x, pdf, color='purple')
ax[0].set_title('Normal Distribution (PDF)')
ax[0].fill_between(x, pdf, alpha=0.2, color='purple')
ax[0].grid()

ax[1].plot(x, cdf, color='orange')
ax[1].set_title('Cumulative Distribution (CDF)')
ax[1].grid()
plt.show()
```

## 🧠 Connection to Machine Learning

### Naive Bayes Classifier
The Naive Bayes algorithm directly applies Bayes' Theorem for classification. To predict the class of a data point, it calculates the probability of each class given the features $P(Class | Features)$ and chooses the class with the highest probability.

### Probabilistic ML Models
Many models, such as Gaussian Mixture Models or Bayesian Neural Networks, heavily rely on probability distributions (like the Normal distribution) to model the underlying structure of the data and quantify uncertainty in their predictions.
