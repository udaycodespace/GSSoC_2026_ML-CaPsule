# DataFrame Methods

## Data Viewing & Inspection

These methods help us quickly inspect and understand a dataset.

### 1. `head()`
- Shows the first **n** rows of the DataFrame.
- Default value: `n = 5`

### 2. `tail()`
- Shows the last **n** rows of the DataFrame.
- Default value: `n = 5`

### 3. `sample()`
- Shows random **n** rows from the DataFrame.
- Default value: `n = 1`


### 4. `nunique()`
- Returns the count of distinct (unique) values in each column.

---

## Useful DataFrame Attributes

### 1. `shape`
- Returns the dimensions of the DataFrame in the form:
  `(number_of_rows, number_of_columns)`

### 2. `columns`
- Returns the list of column names.



```python
import pandas as pd
data = {
'Name': ['Aarav', 'Isha', 'Rohan', 'Sneha', 'Vikram'],
'Age': [25, 30, 35, 40, 45],
'City': ['Delhi', 'Mumbai', 'Bangalore', 'Kolkata', 'Chennai']
}
df = pd.DataFrame(data)
print(df.head())
print(df.tail())
print(df.sample())
print(df.nunique())
print(df.shape)
print(df.columns)
```

# Indexing and Data Selection

## Column Selection

- Single Column → Series
- Multiple Columns → DataFrame

## Row Selection

### loc

Uses labels.

End index is included.

### iloc

Uses integer positions.

End index is excluded.

### Fast Cell Access

- at
- iat


```python
import pandas as pd

data = {
    "Name": ["Aarav", "Isha", "Rohan"],
    "Age": [25, 30, 35],
    "City": ["Delhi", "Mumbai", "Bangalore"]
}

df = pd.DataFrame(data)

# Column Selection
print(df["City"])

print(df[["City", "Age"]])

# loc
print(df.loc[0])

print(df.loc[0:2])

print(df.loc[0, "City"])

# iloc
print(df.iloc[0])

print(df.iloc[0:2])

print(df.iloc[0, 2])

# at
print(df.at[0, "City"])

# iat
print(df.iat[0, 2])
```

# Filtering and Querying Data

## Boolean Filtering

Operators:

- &
- |
- '>'
- <
- '>='
- <=
- ==
- !=

## Query Method

Provides SQL-like filtering.

Uses string conditions.

Query returns a copy.

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Isha", "Rohan"],
    "Age": [25, 30, 35],
    "City": ["Delhi", "Mumbai", "Bangalore"],
    "Salary": [40000, 60000, 80000],
    "Country": ["IN", "IN", "US"]
}

df = pd.DataFrame(data)

print(df[df["Age"] > 30])

print(
    df[
        (df["Age"] > 30) &
        (df["Salary"] > 50000)
    ]
)

print(
    df.query("Age > 30 and Salary < 90000")
)

country = "IN"

print(
    df.query("Country == @country")
)
```

# Data Cleaning

Data cleaning is the process of fixing missing, duplicate, or incorrect data.

## Missing Values

Methods:

- isnull()
- isnull().sum()
- dropna()
- fillna()
- ffill()
- bfill()

### Purpose

- Detect missing values
- Remove missing values
- Replace missing values
- Fill using previous or next value

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Priya"],
    "Age": [20, np.nan, 22]
})

print(df.isnull())

print(df.isnull().sum())

print(df.dropna())

print(df.fillna(0))

df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)

print(df.ffill())

print(df.bfill())
```

# String Cleaning

Pandas provides .str accessor for string manipulation.

## Methods

- lower()
- upper()
- capitalize()
- strip()
- split()
- contains()

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["aman gupta", "rahul kumar"],
    "Country": ["India", "USA"]
})

print(df["Name"].str.lower())

print(df["Name"].str.upper())

print(df["Name"].str.capitalize())

print(df["Name"].str.split(" "))

print(df["Country"].str.contains("USA"))

print(df["Country"].str.contains("india", case=False))
```

# Transforming Data

Transformations modify existing data or create new data.

## Methods

- apply()
- map()
- assign()
- replace()

Used for:

- Custom functions
- Value mapping
- Creating new columns
- Replacing values

```python
import pandas as pd

df = pd.DataFrame({
    "Age": [20, 25, 30],
    "City": ["Delhi", "Mumbai", "Delhi"]
})

df["Age_plus_10"] = df["Age"].apply(
    lambda x: x + 10
)

print(df)

df["City"] = df["City"].map({
    "Delhi": "DL",
    "Mumbai": "MB"
})

print(df)

df = df.assign(
    Bonus=lambda x: x["Age"] * 100
)

print(df)

df["City"] = df["City"].replace({
    "DL": "Delhi"
})

print(df)
```

# Renaming and Reordering

Used to improve readability.

## Methods

- rename()
- columns
- reset_index()

## Reordering

Columns can be rearranged in any order.

```python
import pandas as pd

df = pd.DataFrame({
    "A": [1,2,3],
    "B": [4,5,6],
    "C": [7,8,9]
})

df_renamed = df.rename(
    columns={
        "A":"X",
        "B":"Y"
    }
)

print(df_renamed)

df.columns = ["X","Y","Z"]

print(df)

df_reordered = df[["Z","X","Y"]]

print(df_reordered)

print(df.reset_index())

print(df.reset_index(drop=True))
```

# Sorting and Ranking

Sorting arranges data.

Ranking assigns position numbers.

## Methods

- sort_values()
- sort_index()

```python
import pandas as pd

df = pd.DataFrame({
    "Age":[25,20,30],
    "Income":[50000,80000,60000]
})

print(df.sort_values("Income"))

print(
    df.sort_values(
        "Income",
        ascending=False
    )
)

print(
    df.sort_values(
        ["Age","Income"]
    )
)

print(df.sort_index())

df["Rank"] = df["Income"].rank(
    ascending=False,
    method="dense"
)

print(df)
```

# GroupBy and Aggregation

GroupBy splits data into groups.

Aggregation summarizes grouped data.

## Aggregation Functions

- sum()
- mean()
- count()
- max()
- min()
- std()
- agg()

```python
import pandas as pd

df = pd.DataFrame({
    "Country":["India","India","USA","USA"],
    "Income":[50000,70000,80000,90000],
    "Age":[20,25,30,35]
})

print(
    df.groupby("Country")["Income"].mean()
)

print(
    df.groupby("Country")["Income"].max()
)

print(
    df.groupby("Country").agg({
        "Age":"mean",
        "Income":"mean"
    })
)

print(
    df.groupby("Country")["Income"].agg(
        ["mean","min","max"]
    )
)
```

# Melt

Converts wide format into long format.

## Parameters

- id_vars
- value_vars
- var_name
- value_name

```python
import pandas as pd

df = pd.DataFrame({
    "Country":["USA","USA","India","India"],
    "Year":[2020,2021,2020,2021],
    "Sales":[100,120,90,110],
    "Profit":[20,25,18,22]
})

melted = df.melt(
    id_vars=["Country","Year"],
    value_vars=["Sales","Profit"],
    var_name="Metric",
    value_name="Value"
)

print(melted)
```

# Pivot

Converts long format back to wide format.

## Parameters

- index
- columns
- values

```python
pivoted = melted.pivot(
    index=["Country","Year"],
    columns="Metric",
    values="Value"
)

print(pivoted)
```

# Combining Data

## Merge

SQL-like joins.

Types:

- Inner Join
- Left Join
- Right Join
- Outer Join

## Concat

Used to stack DataFrames.

## Join

Combines using index.

```python
import pandas as pd

customers = pd.DataFrame({
    "customer_id":[1,2,3,4],
    "name":["Adam","Bob","Charlie","Dave"]
})

orders = pd.DataFrame({
    "order_id":[101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount":[250,120,300,180]
})

print(
    pd.merge(
        customers,
        orders,
        on="customer_id"
    )
)

print(
    pd.merge(
        customers,
        orders,
        on="customer_id",
        how="left"
    )
)

print(
    pd.merge(
        customers,
        orders,
        on="customer_id",
        how="outer"
    )
)
```

# Concatenation

Used to combine DataFrames vertically or horizontally.

## Methods

- concat()

```python
import pandas as pd

df1 = pd.DataFrame({
    "id":[1,2,3],
    "name":["Adam","Bob","Charlie"]
})

df2 = pd.DataFrame({
    "id":[4,5,6],
    "name":["David","Eva","Frank"]
})

print(pd.concat([df1,df2]))

print(
    pd.concat(
        [df1,df2],
        ignore_index=True
    )
)

print(
    pd.concat(
        [df1,df2],
        axis=1
    )
)
```
